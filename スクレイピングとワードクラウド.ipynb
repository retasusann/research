```python
#それぞれでまとめた値をまた、df_1で表す。
df_1=pd.read_csv("/content/miyazakikannkou.csv")
df_2=pd.read_csv("/content/miyazakiryokou.csv")
#データフレームの結合
df=pd.concat([df_1,df_2])
#重複した文章（キャプション）の削除
df=df.drop_duplicates()

def remove_emoji(text):
#入力が文字列であることを確認。そうでない場合は、からの文字列を返す
  if not isinstance(text, str):
    return ''
  #テキスト内の絵文字をまずは、英語の単語に変換する
  text=emoji.demojize(text)
  #テキスト内から、名詞・動詞・形容詞を抽出
  results=nagisa.extract(text,extract_postags=['名詞','動詞','形容詞'])
  #resultsをリストに格納
  words=results.words
  #新しい文字列を作成。つまり、元のテキストから絵文字と指定した品詞以外の単語が削除される。
  text=''.join(words)
  # ＃（\#）と数字(\d+)を削除
  #  (削除するもの、削除したものがどうなるか、どのテキストを基に)
  text = re.sub(r'\#|\d+','',text)
   # 英語のemoji単語を除去「一つ以上の連続する英語スペルまたは＿（アンダースコア）」
  text = re.sub(r'[a-zA-Z_]+', '', text)
   # 改行文字(\n)をスペースに置き換え
  text = text.replace('\n', ' ')
  #特定の単語の削除
  text = re.sub(r'宮崎|旅行|観光|宮崎県|宮崎市|市|県|し|いる|旅|ある|ください|こと|人|繋がり|あり|する|なり|い','', text)
  #MeCabのTaggerオブジェクトを初期化する。"-Owakati"はMeCabのオプションで、出力が分かち書きされた形式に変換される
  mecab = mc.Tagger("-Owakati")
  #初期化したTaggerオブジェクトのparseメソッドを使用して、テキストを解析（形態素解析）。この結果、テキストは分かち書きされた形式に変換されます。
  text = mecab.parse(text)
  return text

text=df["caption"]
#remove_emoji関数を利用して、textから絵文字と特定の単語の削除
text_1= text.apply(remove_emoji)
#各要素をスペースで連結して
text_2 = ' '.join(text_1.tolist())

#フォントの指定
font_path="/usr/share/fonts/truetype/fonts-japanese-gothic.ttf"
#ワードクラウドの作成
wordcloud = WordCloud(width=800, height=400, background_color='white', font_path=font_path).generate(text_2)

# プロット設定
plt.figure(figsize=(10, 5))
#interpolation='bilinear'で画像補完
#バイリニア補間（Bilinear Interpolation）は、元の画像に存在しない画素値を生成できる
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
